```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('sales_data.csv', parse_dates=['Date'])
```

```python
df = df.drop_duplicates()

# Feature engineering
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()
df['Hour'] = df['Date'].dt.hour
```

```python
print("Null values:\n", df.isnull().sum())
print("\nDataset Overview:\n", df.describe(include='all'))
```

```python
sns.set(style="whitegrid")

# 1. Monthly Revenue Trend
monthly_revenue = df.groupby('Month')['Total Amount'].sum().reindex([
    'January','February','March','April','May','June','July','August','September','October','November','December'
])
plt.figure(figsize=(12, 6))
monthly_revenue.plot(kind='bar', color='teal')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Sales by Product Category
category_sales = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
category_sales.plot(kind='bar', color='coral')
plt.title('Sales by Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Payment Method Distribution
payment_counts = df['Payment Method'].value_counts()
plt.figure(figsize=(6, 6))
payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 4. Revenue by City
city_sales = df.groupby('City')['Total Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
city_sales.plot(kind='barh', color='skyblue')
plt.title('Total Sales by City')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()

# 5. Heatmap of Sales by Day and Hour
heatmap_data = df.groupby(['Day', 'Hour'])['Total Amount'].sum().unstack().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', linewidths=0.5)
plt.title('Sales Heatmap by Day and Hour')
plt.xlabel('Hour')
plt.ylabel('Day')
plt.tight_layout()
plt.show()
```

```python
# 6. Top 10 Products by Revenue
top_products = df.groupby('Product')['Total Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_products.plot(kind='bar', color='orange')
plt.title('Top 10 Products by Revenue')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 7. Top 10 Customers by Revenue
top_customers = df.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_customers.plot(kind='bar', color='purple')
plt.title('Top 10 Customers by Revenue')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 8. Average Basket Size (Quantity per Transaction)
avg_basket_size = df['Quantity'].mean()
print(f"Average Basket Size (Quantity/Transaction): {avg_basket_size:.2f}")

# 9. Average Revenue per Transaction
avg_revenue = df['Total Amount'].mean()
print(f"Average Revenue per Transaction: ₹{avg_revenue:.2f}")

# 10. Daily Sales Trend
df['Date_only'] = df['Date'].dt.date
daily_sales = df.groupby('Date_only')['Total Amount'].sum()
plt.figure(figsize=(14, 6))
daily_sales.plot(color='green', linewidth=1)
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
