```python
# ========================================
# 1️⃣ Install & Import Libraries
# ========================================
!pip install nltk seaborn wordcloud

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# ========================================
# 2️⃣ Load Dataset
# ========================================
from google.colab import files
uploaded = files.upload()

# Replace with your uploaded CSV file name
df = pd.read_csv("tweet_sentiment_10K.csv")

print("Data Shape:", df.shape)
print(df.head())

# ========================================
# 3️⃣ Rename & Map Sentiment Labels
# ========================================
df.rename(columns={'target': 'sentiment'}, inplace=True)
df['sentiment'] = df['sentiment'].map({0: 'negative', 1: 'positive'})

print("\nSentiment counts:")
print(df['sentiment'].value_counts())

# ========================================
# 4️⃣ Text Preprocessing
# ========================================
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = str(text).lower()
    text = nltk.RegexpTokenizer(r"\w+").tokenize(text)
    text = [lemmatizer.lemmatize(word) for word in text if word not in stop_words]
    return " ".join(text)

df['clean_text'] = df['text'].apply(clean_text)

print("\nCleaned Text Sample:")
print(df[['text', 'clean_text']].head())

# ========================================
# 5️⃣ EDA - Sentiment Distribution
# ========================================
plt.figure(figsize=(6,4))
sns.countplot(x='sentiment', data=df, palette="viridis")
plt.title("Sentiment Distribution")
plt.show()

# ========================================
# 6️⃣ WordCloud for Positive & Negative
# ========================================
positive_text = " ".join(df[df['sentiment'] == 'positive']['clean_text'])
negative_text = " ".join(df[df['sentiment'] == 'negative']['clean_text'])

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
WordCloud(width=500, height=400, background_color='white').generate(positive_text).to_image().show()
plt.title("Positive Sentiment WordCloud")

plt.subplot(1,2,2)
WordCloud(width=500, height=400, background_color='black').generate(negative_text).to_image().show()
plt.title("Negative Sentiment WordCloud")

# ========================================
# 7️⃣ Sentiment Scoring with TextBlob
# ========================================
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

df['polarity'] = df['clean_text'].apply(get_polarity)

# ========================================
# 8️⃣ Polarity Distribution
# ========================================
plt.figure(figsize=(6,4))
sns.histplot(df['polarity'], bins=30, kde=True, color='purple')
plt.title("Polarity Score Distribution")
plt.show()

# ========================================
# 9️⃣ Sentiment Trend Over Time
# ========================================
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    trend_df = df.groupby(df['date'].dt.date)['polarity'].mean()

    plt.figure(figsize=(12,5))
    plt.plot(trend_df.index, trend_df.values, marker='o', color='blue')
    plt.title("Sentiment Trend Over Time")
    plt.xlabel("Date")
    plt.ylabel("Average Polarity")
    plt.xticks(rotation=45)
    plt.show()

# ========================================
# 🔟 Save Processed Data
# ========================================
df.to_csv("processed_sentiment_data.csv", index=False)
print("\nProcessed dataset saved as processed_sentiment_data.csv")
```
